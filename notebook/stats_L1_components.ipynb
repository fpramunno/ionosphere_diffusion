```python
# Sort both DataFrames
df_files = df_files.sort_values("time")
df_clean_2min = df_clean_2min.sort_values("earth_time")
```

```python
import numpy as np
from tqdm import tqdm

t_clean = df_clean_2min["earth_time"].view("int64").to_numpy()
t_maps  = df_files["time"].view("int64").to_numpy()

t_clean
```

```python
#For each arrival time, this tells you where it would be inserted in the map timeline.

t_maps.sort()
df_files = df_files.sort_values("time").reset_index(drop=True)
df_clean_2min = df_clean_2min.sort_values("earth_time").reset_index(drop=True)

idx = np.searchsorted(t_maps, t_clean)
```

```python
# We look at the timestamps immediately before and after the insertion point:

prev_idx = np.clip(idx - 1, 0, len(t_maps)-1)
next_idx = np.clip(idx, 0, len(t_maps)-1)

diff_prev = np.abs(t_clean - t_maps[prev_idx])
diff_next = np.abs(t_clean - t_maps[next_idx])
```

```python
# Basically this does the following if the the idx is nearer to the previous timestamp, choose that, if it's nearer to the next timestamp, choose that, otherwise (tie) choose randomly between the two.

choice = np.where(
    diff_prev < diff_next, prev_idx,
    np.where(diff_next < diff_prev, next_idx,
             np.random.choice([0,1], size=len(t_clean))))  # random tie-breaker
```

```python
# â€œFor each Earth arrival, keep only the first arrival that chose each map; if multiple arrivals picked the same map, drop the later ones.â€

pairs = np.column_stack((np.arange(len(t_clean)), choice))
# Drop duplicates on map index, keep first (or random)
_, unique_idx = np.unique(pairs[:,1], return_index=True)
pairs_unique = pairs[np.sort(unique_idx)]


# If we do the code underneath insetead we ranodmize the order instead of alway taking the first one it randomly takes an index among the ones associated to the same map

# unique_maps = np.unique(pairs[:,1])
# unique_idx = []
# for m in unique_maps:
#     candidates = np.where(pairs[:,1] == m)[0]
#     unique_idx.append(np.random.choice(candidates))
# pairs_unique = pairs[np.sort(unique_idx)]
```

```python
df_matched_unique = pd.DataFrame({
    "earth_time": df_clean_2min.loc[pairs_unique[:,0], "earth_time"].values,
    "time_map":   df_files.loc[pairs_unique[:,1], "time"].values
})
df_matched_unique["dt_diff"] = (
    df_matched_unique["time_map"] - df_matched_unique["earth_time"]
)

```

```python
import numpy as np
import os
import pandas as pd
from scipy.stats import kurtosis, skew
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm

base_dir = "/mnt/nas05/data01/francesco/sdo_img2img/sde_mag2mag_v2/progetto_simone/data/pickled_maps"
file_list = sorted([
    os.path.join(base_dir, f)
    for f in os.listdir(base_dir)
    if f.endswith(".npy")
])

def process_file(path):
    try:
        arr = np.load(path, allow_pickle=True, mmap_mode='r')
        return {
            "filename": os.path.basename(path),
            "min": np.nanmin(arr[0]),
            "max": np.nanmax(arr[0]),
            "mean": np.nanmean(arr[0]),
            "skewness": skew(arr[0].ravel(), nan_policy='omit'),
            "kurtosis": kurtosis(arr[0].ravel(), nan_policy='omit'),
        }
    except Exception as e:
        # Immediately crash the whole program
        print(f"\nâŒ Error while processing {path}: {e}")
        raise

stats = []

with ThreadPoolExecutor(max_workers=os.cpu_count()) as executor:
    futures = [executor.submit(process_file, path) for path in file_list]
    for f in tqdm(as_completed(futures), total=len(futures)):
        # This will raise immediately if any thread hit an exception
        stats.append(f.result())

df = pd.DataFrame(stats)
print(df)

```

```python
import numpy as np
import json
from pathlib import Path
from tqdm import tqdm
from multiprocessing import Pool, cpu_count

def compute_file_stats(file_path):
    """Compute mean and std for a single file."""
    try:
        data = np.load(file_path, allow_pickle=True)
        flat_data = data[0].flatten()
        return {
            'file': file_path.name,
            'mean': float(np.mean(flat_data)),
            'std': float(np.std(flat_data))
        }
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return None

# Compute stats on entire dataset
data_path = Path('/mnt/nas05/data01/francesco/sdo_img2img/sde_mag2mag_v2/progetto_simone/data/pickled_maps/')
all_files = sorted(list(data_path.glob("*.npy")))

print(f"Found {len(all_files)} files")

num_workers = cpu_count() - 1
print(f"Using {num_workers} workers")

# Compute stats in parallel
with Pool(num_workers) as pool:
    results = list(tqdm(
        pool.imap(compute_file_stats, all_files, chunksize=100),
        total=len(all_files),
        desc="Computing per-file stats"
    ))

# Filter out None values
stats = [r for r in results if r is not None]
# with open("normalization_stats.json", "w") as f:
#     json.dump(stats, f)

#print(f"Mean: {mean}, Std: {std}")
```

```python
import json
import numpy as np

# Load your existing per-file stats
with open('/mnt/nas05/data01/francesco/progetto_simone/per_file_normalization_stats.json', 'r') as f:
    per_file_stats = json.load(f)

# Compute global mean by averaging all file means (weighted by data size)
# Each file has 24 * 360 = 8640 values
file_size = 24 * 360

total_sum = sum(s['mean'] * file_size for s in per_file_stats)
total_count = len(per_file_stats) * file_size
global_mean = total_sum / total_count

# For global std, we need to use variance formula
# Var(X) = E[X^2] - E[X]^2
total_var_sum = sum((s['std']**2 + s['mean']**2) * file_size for s in per_file_stats)
global_var = total_var_sum / total_count - global_mean**2
global_std = np.sqrt(global_var)

global_stats = {
    "mean": float(global_mean),
    "std": float(global_std)
}

# Save it
with open('/mnt/nas05/data01/francesco/progetto_simone/global_normalization_stats.json', 'w') as f:
    json.dump(global_stats, f, indent=2)

print(f"Global mean: {global_mean}")
print(f"Global std: {global_std}")
```
